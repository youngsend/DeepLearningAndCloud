## adversarial exampleを生成する方法１：Fast Sign Gradient Method (FSGM)
- Adversarial trainingは以前AI勉強会で紹介した。https://github.com/youngsend/DeepLearningAndKeras/blob/master/Deep-Learning-Ian-Goodfellow/Chapter7_Regularization.pdf
 - 「Adversarial（敵対）Training:背景や原因、対策」ページ。
- FSGM方法は結構簡単。ただ下記の式（名前の通り）です。
 - $x+\epsilon sign(\nabla_x J(\theta,x,y))$.
 - 注意点：普段は$\theta$に対してgradientを計算するが、ここは入力画像$x$に対してgradientを計算している。
